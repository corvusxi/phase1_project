```python
import pandas as pd
import datetime
```

```python
pd.set_option("display.max_columns", None)
```

```python
# Read the CSV
airplane_df = pd.read_csv("../data/airplane_df.csv", encoding = "latin-1", low_memory = False)
```

```python
# United States (and its territories) only
airplane_df = airplane_df[airplane_df["Country"] == "United States"]
```

Analyzing Aircraft Damage

```python
airplane_df["Aircraft.damage"].value_counts()
```

```python
# Drop null values in Aircraft.damage column
airplane_df.dropna(subset=["Aircraft.damage"], inplace = True)
```

Analyzing data if Unkwown should be dropped

```python
unknown_aircraft_damage = airplane_df[airplane_df["Aircraft.damage"] == "Unknown"]
unknown_aircraft_damage.sort_values("Total.Uninjured", ascending=False)
unknown_aircraft_damage
```

Conclusion: Drop Unknown, not that many data points are that signifacnt, also considering there are only 43 values in this category.

```python
airplane_df = airplane_df.drop(airplane_df[airplane_df["Aircraft.damage"] == "Unknown"].index)
```

Analyzing injuries on Destroyed Aircraft, followed by reading investigation reports on the NTSB for clarification on the accident from human error rather than mechanical error.

```python
destroyed_aircraft_damage = airplane_df[airplane_df["Aircraft.damage"] == "Destroyed"]
```

```python
destroyed_aircraft_damage.sort_values("Total.Fatal.Injuries", ascending=False).head(10)
```

```python
destroyed_aircraft_damage.sort_values("Total.Serious.Injuries", ascending=False).head(10)
```

```python
destroyed_aircraft_damage.sort_values("Total.Minor.Injuries", ascending=False).head(10)
```

```python
destroyed_aircraft_damage.sort_values("Total.Uninjured", ascending=False).head(10)
```

```python
# Drop Purpose of Flight Skydiving
destroyed_aircraft_damage = destroyed_aircraft_damage.drop(destroyed_aircraft_damage[destroyed_aircraft_damage["Purpose.of.flight"] == "Skydiving"].index)
```

```python
destroyed_aircraft_damage.sort_values("Total.Fatal.Injuries", ascending=False).head(10)
```

Analyzing data after January 1998. This date was chosen due to the founding of the Commercial Aviation Safety Team (CAST) in the US, who had a goal to reduce the commercial aviation fatality rate in the country.

```python
# Analyzing data after January 1998
destroyed_aircraft_damage["Event.Date"] = pd.to_datetime(destroyed_aircraft_damage["Event.Date"])
eventdate_destroyed_aircraft_damage = destroyed_aircraft_damage[destroyed_aircraft_damage ["Event.Date"] > "1998-01-01"]
fatal_eventdate_destroyed_aircraft_damage = eventdate_destroyed_aircraft_damage.sort_values("Total.Fatal.Injuries", ascending=False).head(10)
fatal_eventdate_destroyed_aircraft_damage
```
